```python
import pandas as pd
import numpy as np
```

```python
df_feb = pd.read_csv('data/feb-inter.csv')
df_mar = pd.read_csv('data/march-inter.csv')

# Unimos estos datasets con los que incluyen los casos confirmados a día de hoy y la población
df_feb2 = pd.read_csv('data/feb2.csv')
df_mar2 = pd.read_csv('data/march2.csv')

df_feb = df_feb.merge(df_feb2)
df_mar = df_mar.merge(df_mar2)

df = pd.concat([df_feb,df_mar],ignore_index=True)
# Creamos una estimación de casos entrantes desde un cierto país a otro cada día
df['CasesEstimation'] = np.divide(np.multiply(df['NumPassengers'],df['OriginConfirmed'], dtype=np.double),
                                  df['OriginPopulation'], dtype=np.double)
df.head()
```

```python
# Agrupamos las estimaciones de contagios entrantes por país de llegada y fecha
new_df = df.groupby(['Zone','Date','ConfirmedCasesWeekLater','ConfirmedCases','Population']).sum()[['CasesEstimation','ImportedRisk']].reset_index()
```

```python
new_df = new_df[['Zone', 'Date', 'ConfirmedCases','CasesEstimation','ImportedRisk', 'ConfirmedCasesWeekLater','Population']]
new_df.head()
```

```python
new_df.to_csv('data/inter.csv')
```
